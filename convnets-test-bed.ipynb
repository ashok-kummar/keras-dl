```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.utils import load_img, img_to_array, array_to_img

import numpy as np
import matplotlib.pyplot as plt
import os
import random

tf.config.list_physical_devices()
```

```python
plt.axis("off")
plt.imshow(load_img("car1.jpg"))
```

```python
img = img_to_array(load_img("car1.jpg"))
img = np.expand_dims(img, axis=0)
print("shape of input img = ", img.shape)
```

```python
print("shape of input img = ", img.shape)

print("conv info: filters=32, kernel=3, strides=1, padding=same")
x = layers.Conv2D(32, 3, strides=1, padding="same")(img)

x_shape = x.shape
print("shape of conv output = ", x.shape)

rows = 2
cols = 5
fig=plt.figure(figsize=(20,10))
for i in range(rows*cols):
    fig.add_subplot(rows, cols, i+1)
    image = array_to_img(np.reshape(x[0, :, :, i],(x_shape[1],x_shape[2],1)))
    plt.imshow(image, cmap="gist_gray")
    plt.axis("off")

    
fig.tight_layout()    
plt.show()
```
